```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Let randomly create some dataâ€™s for two variables,

x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)
#Visualize the variables spreads for better understanding

x = x[:, np.newaxis] 
y = y[:, np.newaxis] 

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.show()
```

```python
import sklearn.metrics as metrics
mse = metrics.mean_squared_error(x,y)
rmse = np.sqrt(mse) 
r2 = metrics.r2_score(x,y)
print('RMSE value:',rmse)
print('R2 value:',r2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_features1 = PolynomialFeatures(degree=2)
x_poly1 = polynomial_features1.fit_transform(x)
model1 = LinearRegression()
model1.fit(x_poly1, y)
y_poly_pred1 = model1.predict(x_poly1)
from sklearn.metrics import mean_squared_error, r2_score
rmse1 = np.sqrt(mean_squared_error(y,y_poly_pred1))
r21 = r2_score(y,y_poly_pred1)
print(rmse1)
print(r21)
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error

# Generate sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 5, 10, 17, 26])

# Reshape the data
X = x.reshape(-1, 1)
Y = y.reshape(-1, 1)

# Create a polynomial features object of degree 2
poly = PolynomialFeatures(degree=2)

# Transform the features to include polynomial terms up to degree 2
X_poly = poly.fit_transform(X)

# Create a linear regression object
model = LinearRegression()

# Fit the model using the transformed features
model.fit(X_poly, Y)

# Generate predictions
y_pred = model.predict(X_poly)

# Evaluate the model using R-squared and RMSE
r2 = r2_score(Y, y_pred)
rmse = np.sqrt(mean_squared_error(Y, y_pred))

# Plot the data and the predicted polynomial function
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.title(f"Polynomial Regression\nR-squared: {r2:.2f}, RMSE: {rmse:.2f}")
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error

# Generate random data
np.random.seed(42)
x1 = np.random.uniform(low=-5, high=5, size=100)
x2 = np.random.uniform(low=-5, high=5, size=100)
y = np.exp(-x1**2/4) + np.sin(x2) + np.random.randn(100)*0.1

# Reshape the data
X = np.vstack([x1, x2]).T
Y = y.reshape(-1, 1)

# Create a polynomial features object of degree 2
poly = PolynomialFeatures(degree=2)

# Transform the features to include polynomial terms up to degree 2
X_poly = poly.fit_transform(X)

# Create a linear regression object
model = LinearRegression()

# Fit the model using the transformed features
model.fit(X_poly, Y)

# Generate predictions
y_pred = model.predict(X_poly)

# Evaluate the model using R-squared and RMSE
r2 = r2_score(Y, y_pred)
rmse = np.sqrt(mean_squared_error(Y, y_pred))

# Plot the data and the predicted polynomial function
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x1, x2, y)
ax.plot_trisurf(x1, x2, y_pred.ravel(), alpha=0.5)
ax.set_title(f"Polynomial Regression\nR-squared: {r2:.2f}, RMSE: {rmse:.2f}")
plt.show()
```
